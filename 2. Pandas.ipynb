<b> Interpolation </b>

Interpolation in pandas is a way to fill in missing or NaN (Not a Number) values in a DataFrame or Series with estimated values based on the existing data. When it comes to why interpolation might not be working as expected for the first row in pandas, there could be a few different reasons:

Method Selection: Pandas provides several interpolation methods such as 'linear', 'nearest', 'zero', etc. The default method is usually 'linear'. Depending on the nature of your data, the chosen method might not be appropriate for the first row. For example, the 'linear' method may not work well for the first row if the data is not evenly spaced or if the first row has missing values on both sides.

Insufficient Data: Interpolation works by estimating missing values based on neighboring data points. If the first row has missing values and there are no valid neighboring points to use for interpolation, it might not be possible to generate a meaningful interpolated value.

Index Values: The index values of your DataFrame or Series play a significant role in interpolation. If your index values are not numeric or are not in a sequence that can be used for interpolation, the interpolation might not work as expected. Ensure that your index values are suitable for interpolation.

Interpolation Limitations: Interpolation methods have their limitations, and they might not always provide accurate results, especially if the data has complex patterns or irregularities. It's important to understand the characteristics of your data and choose an appropriate interpolation method.

<b> Merge </b>

```python
import pandas as pd
#Merging two data frame means there should be atleast one common column
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]}) #variable C must be diff var name not B

pd.merge(var1,var2,on="A") #we must mention common part
```

```python
# for C column to come before B
pd.merge(var2,var1,on="A")
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]}) #changing one value of A in var2 now only common rows will be displayed
pd.merge(var2,var1,on="A")
```

```python
#Now without on also it works because A is not fully same so it merges only similar parts
pd.merge(var1,var2)
```

```python
#Inner merge
pd.merge(var1,var2,how="inner")  #Usual stuff
```

```python
pd.merge(var1,var2,how="left") # works based on data of left variable
```

```python
pd.merge(var1,var2,how="right") # works based on data of right variable
```

```python
pd.merge(var1,var2,how="outer") #Shows both the data with NaN in missing values
```

```python
#Use indicators to show which all data will be available in both, left and right
pd.merge(var1,var2,how="outer",indicator=True)
```

```python
import pandas as pd
var = pd.read_csv("C:\\Users\\viswa\\Data Science and ML\\Datasets\\Iris.csv")
var
```

```python
#It checks row by row if it find duplicate values then it will remove
#If i try to use it in SepalLengthCm only single same value occurs duplicate values goes off
var.drop_duplicates(subset=["SepalLengthCm"])
```

```python
#we have a parameter keep using which u can have either first one or last one which is occuring
#by default it keeps first one
var.drop_duplicates(subset=["SepalLengthCm"],keep="last")
```

```python
#more than one columns i.e it removes only when a row's together value is occuring again
var.drop_duplicates(subset=["SepalLengthCm","SepalWidthCm"],keep="last")
```
