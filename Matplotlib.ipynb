```python
#Without Overlap
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[35,20,27,72]

width=0.2
p=np.arange(len(x)) # 0 to len-1
#Now in axis we will have 0 0.5 etc..
#but after this also overlap will exist
#so we create p1 where we add elements of p with width
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.")
plt.title("Bar Graph")

plt.bar(p,y,width=0.2,color="r",label="popularity")
plt.bar(p1,z,width=0.2,color="y",label="popularity1")
plt.legend() #only on mentioning legend, label produces output
plt.show()
```

In the context of the code you provided, the gca() function is used to retrieve the current Axes instance of the current figure in Matplotlib. Let's break down the code and explain how the gca() function works:

<b>plt.pie(x, labels=y, radius=1, colors=c):</b> This line of code creates a pie chart using the Matplotlib library. It takes the following parameters:

<b>x:</b> The data values for the pie chart slices.
labels: Labels for each slice of the pie.
radius: The radius of the pie chart. Here, it's set to 1, meaning the pie chart will be drawn within a circle of radius 1.
colors: The colors of the pie chart slices.
cr = plt.Circle(xy=(0,0), radius=0.7, facecolor="w"): This line creates a white circle (circle patch) using the Circle class from Matplotlib's patches module. It takes the following parameters:

<b>xy: </b>The center coordinates of the circle. Here, it's (0, 0), which places the center at the origin of the plot.
radius: The radius of the circle. Here, it's set to 0.7.
facecolor: The color to fill the circle with. Here, it's set to "w", which represents white.
plt.gca().add_artist(cr): This line of code adds the white circle (created using the cr variable) as an artist to the current Axes instance. Here's what's happening step by step:

<b>plt.gca():</b> The gca() function stands for "get current axes." It retrieves the current Axes instance associated with the current figure.
.add_artist(cr): The add_artist() method of the Axes instance is used to add graphical artists, such as the white circle, to the plot. In this case, the white circle created earlier is added to the plot using this method.
Overall, the purpose of adding the white circle to the plot seems to be creating a circular boundary around the pie chart to give it a visual separation from the rest of the plot area. The white circle acts as a background or frame for the pie chart, enhancing its appearance and aesthetics.

# STEM PLOT

A stem plot, also known as a stem-and-leaf plot, is a graphical representation used for displaying a dataset's distribution of values. It is particularly useful for showing the shape of the data, its central tendency, and its spread. Stem plots are often used in exploratory data analysis to quickly visualize the distribution of a dataset, especially when dealing with small to moderately sized datasets.

Here's how a stem plot works:

Each data point is split into a "stem" and a "leaf."
The "stem" typically consists of the leading digits of the data points.
The "leaf" consists of the trailing digits of the data points.

# BOX vs WHISKER PLOT

A box plot, also known as a whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset, including the median, quartiles, and potential outliers.
<br/><br/>
<b><i>Yellow line is median line</b></i>
<br/><br/>
 In a box plot, the terms "first quartile," "second quartile," and "third quartile" correspond to different quantiles of the dataset, and the "median" line also plays a significant role in summarizing the distribution. Let's break down what each of these terms means:
<br/><br/><br/>
<b>First Quartile (Q1):</b>
The first quartile, often denoted as Q1, represents the 25th percentile of the dataset. It is the value below which 25% of the data falls. In other words, if you were to sort the data in ascending order, Q1 would be the value that separates the lowest 25% of the data from the rest.
<br/><br/>
<b>Second Quartile (Q2) - Median: </b>
The second quartile, also known as Q2 or the median, represents the middle value of the dataset when it is ordered. It divides the data into two halves: the lower 50% and the upper 50%. In a box plot, the median line is often shown as a horizontal line inside the box. It's a measure of central tendency and gives you an idea of the "typical" value in the dataset.
<br/><br/>
<b>Third Quartile (Q3): </b>
The third quartile, denoted as Q3, represents the 75th percentile of the dataset. It is the value below which 75% of the data falls. Similar to Q1, Q3 helps you understand how the data is distributed in the upper 75% of the range.
<br/><br/>
Now, let's visualize these quartiles and the median using a box plot. In a box plot:
<br/>
The "box" itself represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3). The width of the box gives you an idea of the spread of the middle 50% of the data.
<br/><br/>
The horizontal line inside the box is the median (Q2). It gives you a sense of the central tendency of the dataset.
<br/><br/>
The "whiskers" extend from the edges of the box to show the range of the data outside the IQR. Any data points beyond the whiskers are considered potential outliers.
<br/><br/>
Potential outliers, which are data points that fall beyond the whiskers, are sometimes represented as individual points in the plot.
<br/><br/>
<h3> CONCEPT OF OUTLIERS IN DETAIL ALONG WITH WHAT IS WHISKERS </h3>

In a box plot, the <b>"whiskers"</b> are lines that extend from the edges of the box to indicate the range of data points beyond the interquartile range (IQR). The whiskers provide insights into the spread of the data and help identify potential outliers in the dataset.
<br/><br/>
<b>The interquartile range (IQR)</b> is the range between the first quartile (Q1) and the third quartile (Q3) of the dataset. The box in the box plot represents this range, and it contains the middle 50% of the data.
<br/><br/>
Here's how whiskers are typically used in a box plot:
<br/><br/>
<b>Upper Whisker:</b>
The upper whisker extends from the top of the box (Q3) to the highest data point within a certain distance from Q3. This distance is often defined as 1.5 times the IQR. Data points beyond the upper whisker are considered potential outliers.
<br/><br/>
<b>Lower Whisker:</b>
The lower whisker extends from the bottom of the box (Q1) to the lowest data point within a certain distance from Q1 (again, usually 1.5 times the IQR). Data points beyond the lower whisker are also potential outliers.
<br/><br/>
<b>Outliers:</b>
Individual data points that fall beyond the whiskers are labeled as potential outliers. These are values that are significantly far from the central distribution of the data and may indicate unusual or anomalous observations.
<br/><br/>
In some variations of box plots, the whiskers might extend to a certain percentile instead of being determined by a fixed distance from Q1 and Q3. This approach can handle data that is skewed or has extreme values more effectively.

In the context of a stack plot, which is also known as a stacked area plot, "baseline," "sym," and "wiggle" are different ways to visualize the individual components or layers that make up the stack. A stack plot is used to show how different data series contribute to a whole while maintaining a sense of their individual values over a common X-axis range.

Let's explore the meanings of these terms in the context of a stack plot:

Baseline (default behavior):
In a stack plot, the baseline is usually set to zero. This means that the layers are stacked on top of each other starting from the zero line on the Y-axis. The area between the baseline and the top of each layer represents the contribution of that data series to the total. The baseline is the reference point from which the other layers are built.

Symmetric ("sym") Stacking:
In symmetric stacking, the layers are stacked on top of each other in a way that their centerlines align with the baseline. This method can make it easier to compare the contributions of different layers across the X-axis. However, it can lead to distortion in the representation of individual values within each layer, especially if some layers have significant negative values.

Wiggle Stacking ("wiggle"):
Wiggle stacking is a variation that attempts to address the distortion issue introduced by symmetric stacking. In wiggle stacking, each layer's centerline does not necessarily align with the baseline. Instead, the layers are shifted vertically to reduce the visual distortion. This allows for a more accurate representation of individual values within each layer, especially when negative values are present. However, it can still lead to some overlapping and complexity in the visualization.

<br/><i> Note: In the context of data visualization, "distortion" refers to a misrepresentation or alteration of the original data when it is presented graphically. </i>

Choosing between baseline, symmetric, or wiggle stacking depends on the context of your data and the specific insights you want to convey:

<b> SUMMARY </b><br/><br/>
Baseline stacking is common when the focus is on showing the overall contribution of each component to the whole.

Symmetric stacking can help emphasize the differences between the components' contributions more clearly, but it might compromise the accuracy of individual values in some cases.

Wiggle stacking aims to strike a balance between emphasizing differences and accurately representing individual values
